### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

def plot_square_function():
    try:
        # Generate x values
        x = np.arange(11)
        
        # Ensure x is a valid array
        if not isinstance(x, np.ndarray):
            raise TypeError("x must be a NumPy array.")
        
        # Ensure x is not empty
        if x.size == 0:
            raise ValueError("Generated x values are empty.")
        
        # Compute y values (handling potential overflow)
        with np.errstate(over='raise'):
            try:
                y = x**2
            except FloatingPointError:
                raise ValueError("Overflow occurred while computing x^2.")

        # Ensure y values are valid
        if not isinstance(y, np.ndarray) or y.size == 0:
            raise ValueError("Generated y values are invalid or empty.")
        
        # Create the plot
        plt.figure(figsize=(8, 5))
        plt.plot(x, y, 'bo-', label='y = x^2')

        # Set plot labels and title
        plt.title("Square Function Plot")
        plt.xlabel('X Values')
        plt.ylabel('Y Values (X squared)')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.legend()

        # Display the plot
        plt.show()

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except TypeError as te:
        print(f"TypeError: {te}")
    except FloatingPointError as fe:
        print(f"FloatingPointError: {fe}")
    except plt.MatplotlibDeprecationWarning as mdw:
        print(f"MatplotlibDeprecationWarning: {mdw}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Call the function
plot_square_function()
```

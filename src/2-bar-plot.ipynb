### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

def plot_category_distribution(categories, values, colors):
    try:
        # Validate inputs
        if not categories or not values:
            raise ValueError("Category and value lists cannot be empty.")
        
        if len(categories) != len(values):
            raise ValueError("Categories and values must have the same length.")
        
        if colors and len(colors) != len(categories):
            raise ValueError("Colors list must match categories in length.")

        # Plot
        plt.figure(figsize=(6, 4))
        plt.bar(categories, values, color=colors)

        # Customization
        plt.title('Category Value Distribution')
        plt.xlabel('Categories')
        plt.ylabel('Values')
        plt.grid(axis='y', linestyle='--', alpha=0.6)

        # Show plot
        plt.show()

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Execute the function
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

plot_category_distribution(categories, values, colors)
```

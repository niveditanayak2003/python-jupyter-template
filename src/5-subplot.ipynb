### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np
# Creating a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Data for each plot
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['skyblue', 'orange', 'green', 'purple']

# Line Plot (Top Left)
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar Plot (Top Right)
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Scatter Plot (Bottom Left)
axes[1, 0].scatter(x, y, color='red', marker='o', label='Data Points')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')
axes[1, 0].grid(True, linestyle='--', alpha=0.6)
axes[1, 0].legend()

# Pie Chart (Bottom Right)
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

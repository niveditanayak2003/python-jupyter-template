### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

def plot_multiple_subplots():
    try:
        fig,axes=plt.subplots(2,2,figsize=(10,8))
        x=np.arange(1,6)
        y=x*2

        categories=['A','B','C','D']
        values=[5,7,3,9]
        labels=['Python','Java','C++','JavaScript']
        sizes=[40,25,20,15]
        colors=['skyblue','orange','green','purple']
        if not (categories and values and labels and sizes):
            raise ValueError("Input lists cannot be empty.")
        
        if len(categories) != len(values):
            raise ValueError("Categories and values must have the same length.")

        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must have the same length.")

        if colors and len(colors) != len(labels):
            raise ValueError("Colors list must match labels in length.")
        
        axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')
        axes[0, 0].set(title='Line Plot', xlabel='X', ylabel='Y')
        axes[0, 0].legend()
        axes[0, 0].grid(True, linestyle='--', alpha=0.6)

        axes[0, 1].bar(categories, values, color=colors[:len(categories)])
        axes[0, 1].set(title='Bar Chart', xlabel='Categories', ylabel='Values')
        axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

        axes[1, 0].scatter(x, y, color='red', marker='o', label='Data Points')
        axes[1, 0].set(title='Scatter Plot', xlabel='X', ylabel='Y')
        axes[1, 0].grid(True, linestyle='--', alpha=0.6)
        axes[1, 0].legend()

        axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', 
                       colors=colors if colors else plt.cm.Paired.colors, 
                       startangle=140, wedgeprops={'edgecolor': 'black'})
        axes[1, 1].set_title('Pie Chart')

        plt.tight_layout()
        plt.show()
    except ValueError as ve:
        print(f"ValueError:{ve}")
    except Exception as e:
        print(f"An unexpected error occurred:{e}")


plot_multiple_subplots()


```

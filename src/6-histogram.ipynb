### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def plot_histogram(data, bins=5):
    try:
        # Validate input data
        if not data:
            raise ValueError("Data list cannot be empty.")

        if not all(isinstance(i, (int, float)) for i in data):
            raise ValueError("All elements in data must be numeric.")

        if not isinstance(bins, int) or bins <= 0:
            raise ValueError("Bins must be a positive integer.")

        # Plot
        plt.figure(figsize=(6, 4))
        plt.hist(data, bins=bins, color='cyan', edgecolor='black', alpha=0.7)

        # Customization
        plt.title('Histogram of Data')
        plt.xlabel('Values')
        plt.ylabel('Frequency')
        plt.grid(axis='y', linestyle='--', alpha=0.6)

        # Show plot
        plt.show()

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Execute the function
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plot_histogram(data,bins=5)
```

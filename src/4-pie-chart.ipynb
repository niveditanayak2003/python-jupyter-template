### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt
def plot_pie_chart(labels,sizes,colors=None):
    try:
        if not(labels and sizes):
            raise ValueError("Labels and sizes lists cannot be empty")
        if len(labels)!=len(labels):
            raise ValueError("labels and sizes must have the same length")
        if colors and len(colors)!=len(labels):
            raise ValueError("Colors list must match the labelsin length.")
        
        plt.figure(figsize=(6,6))
        plt.pie(sizes,labels=labels,autopct='%1.1f%%',colors=colors if colors else plt.cm.Paired.colors,startangle=140,wedgeprops={'edgecolor':'black'})

        plt.title("Programming language Popularity")

        plt.show()
    except ValueError as ve:
        print(f"ValueError:{ve}")
    except Exception as e:
        print(f"An unexpected error occurred:{e}")

labels=['Python','Java','C++','Javascript']
sizes=[40,25,20,15]
colors=['skyblue','orange','green','purple']

plot_pie_chart(labels,sizes,colors)
        
```
